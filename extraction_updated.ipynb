```python
import os
import shutil

# Source directory containing case folders
src_dir = r"new_data_extraction"
# Destination directory to collect all imaging files
dst_dir = r"collected_imaging"
# Create destination directory for segmentations
segmentation_dst_dir = r"collected_segmentations"
os.makedirs(dst_dir, exist_ok=True)
os.makedirs(segmentation_dst_dir, exist_ok=True)

# Loop through all subfolders from 1 to 20 and move segmentation files
for i in range(1, 21):
    folder_name = f"case_{i:02d}"
    folder_path = os.path.join(src_dir, folder_name)
    if os.path.isdir(folder_path):
        src_file = os.path.join(folder_path, "segmentation.nii.gz")
        if os.path.exists(src_file):
            # file should be named as segmentation_case_0000.nii.gz
            dst_file = os.path.join(segmentation_dst_dir, f"segmentation_{i:02d}.nii.gz")
            shutil.copy2(src_file, dst_file)
            print(f"Copied: {src_file} → {dst_file}")
        else:
            print(f"Warning: segmentation.nii.gz not found in {folder_path}")

# Loop through all subfolders from 1 to 20 and move imaging files
for i in range(1, 21):
    folder_name = f"case_{i:02d}"
    folder_path = os.path.join(src_dir, folder_name)
    if os.path.isdir(folder_path):
        src_file = os.path.join(folder_path, "imaging.nii.gz")
        if os.path.exists(src_file):
            case_number = folder_name.split("_")[-1]
            # file should be named as imaging_case_0000.nii.gz
            dst_file = os.path.join(dst_dir, f"imaging_{case_number}_0000.nii.gz")
            shutil.copy2(src_file, dst_file)
            print(f"Copied: {src_file} → {dst_file}")
        else:
            print(f"Warning: imaging.nii.gz not found in {folder_path}")

print("✅ All available imaging files have been collected successfully!")
```

```python
import os

# === Configuration ===
base_path = r"new_data_extraction"  # change path if needed
start_index = 0
end_index = 589   # since 00588 is inclusive

# === Get all existing case folders ===
existing_folders = set(
    f for f in os.listdir(base_path)
    if f.startswith("case_") and os.path.isdir(os.path.join(base_path, f))
)

# === Generate the expected case folder names ===
expected_folders = {f"case_{i:05d}" for i in range(start_index, end_index)}

# === Find missing ones ===
missing_folders = sorted(expected_folders - existing_folders)

# === Output ===
if missing_folders:
    print(f"⚠️ Missing {len(missing_folders)} cases:")
    for folder in missing_folders:
        print(folder)
else:
    print("✅ All case folders from 00000 to 00588 are present!")
```

```python
import os
import shutil

# Source directory containing case folders
src_dir = r"new_data_extraction"

# Destination directories
dst_imaging = r"collected_imaging"
dst_segmentation = r"collected_segmentations"

os.makedirs(dst_imaging, exist_ok=True)
os.makedirs(dst_segmentation, exist_ok=True)

# Loop through all subfolders (you can change range as needed)
for i in range(1, 21):
    folder_name = f"case_{i:05d}"  # case_001, case_002, ...
    folder_path = os.path.join(src_dir, folder_name)

    if os.path.isdir(folder_path):
        # === Copy segmentation ===
        seg_src = os.path.join(folder_path, "segmentation.nii.gz")
        if os.path.exists(seg_src):
            seg_dst = os.path.join(dst_segmentation, f"segmentation_{i:03d}.nii.gz")
            shutil.copy2(seg_src, seg_dst)
            print(f"Copied segmentation → {seg_dst}")
        else:
            print(f"⚠️ segmentation.nii.gz not found in {folder_path}")

        # === Copy imaging ===
        img_src = os.path.join(folder_path, "imaging.nii.gz")
        if os.path.exists(img_src):
            img_dst = os.path.join(dst_imaging, f"imaging_{i:03d}_0000.nii.gz")
            shutil.copy2(img_src, img_dst)
            print(f"Copied imaging → {img_dst}")
        else:
            print(f"⚠️ imaging.nii.gz not found in {folder_path}")

print("✅ All available imaging and segmentation files have been collected successfully!")
```
